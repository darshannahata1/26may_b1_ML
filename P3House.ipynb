```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

df=pd.read_csv('housing.csv')
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms' , 'Area Population']]
y=df['Price']
xtrain , xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=10)
model=LinearRegression()
model.fit(xtrain,ytrain)
model.coef_
pd.DataFrame(model.coef_,columns=['coef_'],index=x.columns)
ypred=model.predict(xtest)
final_model=model
pd.to_pickle(final_model,'HousePricePredictor')
model=pd.read_pickle('HousePricePredictor')

import tkinter as tk
from PIL import ImageTk, Image
app=tk.Tk()
app.geometry("400x250")
app.configure(bg='#34568B')
app.title('House Price Predictior')
c = tk.Canvas(app,width = 400,height = 250)
c.place(x=0,y=0)
img = ImageTk.PhotoImage(Image.open('house.png'))
c.create_image(1,1,anchor=tk.NW, image=img)

income=tk.StringVar()
house_age=tk.StringVar()
number_of_rooms=tk.StringVar()
population=tk.StringVar()

tk.Label(app,text="INCOME",bg='#ffffff',font=('Helvetica', 10, 'bold')).place(x=20,y=40)
tk.Label(app,text="HOUSE AGE",bg='#ffffff',font=('Helvetica', 10, 'bold')).place(x=20,y=70)
tk.Label(app,text="NUMBER OF ROOMS",bg='#ffffff',font=('Helvetica', 10, 'bold')).place(x=20,y=100)
tk.Label(app,text="POPULATION",bg='#ffffff',font=('Helvetica', 10, 'bold')).place(x=20,y=130)

tk.Entry(app,width=30,bg="#ffffff",textvariable=income).place(x=170,y=40)
tk.Entry(app,width=30,bg="#ffffff",textvariable=house_age).place(x=170,y=70)
tk.Entry(app,width=30,bg="#ffffff",textvariable=number_of_rooms).place(x=170,y=100)
tk.Entry(app,width=30,bg="#ffffff",textvariable=population).place(x=170,y=130)


def Predict_price():
    i=income.get()
    
    h=house_age.get()
    n=number_of_rooms.get()
    p=population.get()
    income.set('')
    house_age.set('')
    number_of_rooms.set('')
    population.set('')
    query=pd.DataFrame({'Avg. Area Income':[i],
                    'Avg. Area House Age':[h],
                    'Avg. Area Number of Rooms':[n],
                    'Area Population':[p]
                   })
    
    
    
    Predict_price_var=tk.Variable(app)
    Predict_price_var.set('%.1f$'%(model.predict(query)[0]))
    
    
    Predict_price=tk.Label(app,text='Price for the house can be :',bg='#ADD8E6',font=('Helvetica', 10, 'bold')).place(x=30,y=220)
    Predict_price=tk.Label(app,text ='Price for the house can be :',textvariable=Predict_price_var,bg='#ADD8E6',font=('Helvetica', 10, 'bold')).place(x=220,y=220)
    
tk.Button(app,text='Predict Price',font=('Helvetica', 10, 'bold'),command=Predict_price).place(x=210,y=170)

app.mainloop()
```
